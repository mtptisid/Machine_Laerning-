# Logistic Regression Model 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
%matplotlib inline
```

```python
data = [
    [22,0],
    [25,0],
    [47,1],
    [52,0],
    [46,1],
    [56,1],
    [55,0],
    [60,1],
    [62,1],
    [61,1],
    [18,0],
    [28,0],
    [27,0],
    [29,0],
    [49,1],
    [55,1],
    [25,1],
    [58,1],
    [19,0],
    [18,0],
    [21,0],
    [26,0],
    [40,1],
    [45,1],
    [50,1],
    [54,1],
    [23,0]
]
```

```python
df = pd.DataFrame(data, columns=["age", "bought_insurance"])
```

```python
df
```

```python
df.to_csv("InsuranceDetails.csv", index=False)
```

```python
plt.scatter(df.age,df.bought_insurance,marker='+')
```

```python

from sklearn.model_selection import train_test_split
X_train, X_tes, Y_train, Y_test =train_test_split(df[['age']],df.bought_insurance,test_size=0.2)
```

```python
model = linear_model.LogisticRegression()
```

```python
model.fit(X_train,Y_train)
```

```python
model.coef_
```

```python
model.score(X_tes,Y_test)
```

```python
def predprint(val):
    if model.predict([[val]]) == 0:
        print("No")
    else:
        print('yes')
```

```python
list(map(predprint, Y_test))
```

```python
hrdata.head()
```

```python
ydata = hr_data.left
```

```python
hrX_train, hrX_test, hrY_train, hrY_test = train_test_split(hrdata, ydata, test_size=0.2)
```

```python
hrmodel = linear_model.LogisticRegression()
```

```python
hrmodel.coef_
```
