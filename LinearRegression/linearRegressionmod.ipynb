```python
df = pd.read_csv('canada_data.csv')
```

```python
df.head
df.columns
```

```python
df = df.rename(columns={"year": "Year", "per capita income (US$)": "Income"})

df.columns
```

```python
%matplotlib inline
plt.xlabel('Year')
plt.ylabel('per capita income (US$)')
plt.scatter(df.Year, df.Income)
```

```python
model1 = linear_model.LinearRegression()
model1.fit(df[['Year']],df.Income)
```

```python
model1.predict([[2022]])
```

```python
df.tail
```

```python
yr_list = [2017, 2018, 2019, 2020, 2021, 2022, 2023, 2024, 2025]
pred_list = []
for i in yr_list:
    pred = model1.predict([[i]])
    pred_list.append(f'Prediction for the year {i} is : {pred[0]}')
for i in pred_list:
    print(i)
```

```python
%matplotlib inline
plt.xlabel('Year')
plt.ylabel('per capita income (US$)')
plt.scatter(df.Year, df.Income)
plt.plot(df.Year, model1.predict(df[['Year']]), color='red')
```

# Linear Regression Multiple Variables

```python
# Creating the data as a list of lists
data = [
    [2600, 3, 20, 550000],
    [3000, 4, 15, 565000],
    [3200, None, 18, 610000],
    [3600, 3, 30, 595000],
    [4000, 5, 8, 760000]
]

# Creating the DataFrame
df = pd.DataFrame(data, columns=["area", "bedrooms", "age", "price"])

# Displaying the DataFrame
print(df)
```

```python
import math
df.bedrooms.fillna(math.floor(df.bedrooms.median()))
```

```python
df.bedrooms = df.bedrooms.fillna(math.floor(df.bedrooms.median()))
df
```

### Create a Model with Linear Regression

```python
model2 = linear_model.LinearRegression()
```

```python
model2.fit(df[['area','bedrooms','age']],df.price)
```

```python
model2.coef_
```

```python
model2.intercept_
```

```python
model2.predict([[3000, 3, 40]])
```

```python
data = [
    [None,8,  9, 50000],
    [None,8, 6, 45000],
    ["five", 6, 7, 60000],
    ["two", 10, 10, 65000],
    ["seven", 9, 6, 70000],
    ["three", 7, 10, 62000],
    ["ten", None, 7, 72000],    
    ["eleven", 7, 8, 80000]]

# Creating the DataFrame with column names
df = pd.DataFrame(data, columns=["experience", "test_score(out of 10)", "interview_score(out of 10)", "salary($)"])

# Displaying the DataFrame
print(df)
    
    
    
```

```python
df.experience = df.experience.fillna("zero")
df.experience.apply(w2n.word_to_num)
```

```python
df
```

```python
df.rename(columns={"experience": "experience", "test_score(out of 10)": "test_score", "interview_score(out of 10)": "interview_score", "salary($)": "salary"}, inplace=True)
```

```python
df
```

```python
model3 = linear_model.LinearRegression()
model3.fit(df[['experience', 'test_score' , 'interview_score']], df.salary)
```

```python
model3.coef_
```

```python
model3.intercept_
```
