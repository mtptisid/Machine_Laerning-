## Linear Regression with Pandas dummies variable and OneHotEncoder

### load needed libraries and modules and DF

```python
import pandas as pd
import numpy as np
import matplotlib as plt
from sklearn import linear_model
```

```python
data =[
["monroe township",3000,565000],
["monroe township",3200,610000],
["monroe township",2600,550000],
["monroe township",3600,680000],
["monroe township",4000,725000],
["west windsor",2600,585000],
["west windsor",2800,615000],
["west windsor",3300,650000],
["west windsor",3600,710000],
["robinsville",2600,575000],
["robinsville",2900,600000],
["robinsville",3100,620000],
["robinsville",3600,695000],
]

data
```

```python
df = pd.DataFrame(data, columns=["Town" , "Area", "Price"])
```

```python
df.to_csv("HousePrice.csv", index=False)
df
```

```python
df.info()
```

## #Model with pandas dummie variables

```python
dummies = pd.get_dummies(df["Town"], dtype='int64')
dummies
```

```python
merged = pd.concat([df, dummies], axis='columns')
merged
```

```python
final=merged.drop(["Town", "west windsor"], axis='columns')
```

```python
final
```

```python
X=final.drop("Price", axis='columns')
y=final.Price
X
```

```python
y
```

## Load a model and fit 

```python
model1 = linear_model.LinearRegression()
```

```python
model1.fit(X,y)
```

```python
model1.predict([[3000,0,1]])
```

```python
model1.fit(X.values,y)
```

```python
model1.predict([[3000,0,1]])
```

```python
model1.predict([[3000,0,0]])
```

```python
model1.predict([[2600,0,0]])
```

```python
model1.score(X.values,y)
```

```python
model1.predict([[3400,0,0]])
```

```python
model1.predict([[2800,0,1]])
```

### Since the latest build of sklearn library removed categorical_features parameter for onehotencoder class. It is advised to use ColumnTransformer class for categorical datasets. Refer the sklearn's official documentation for futher clarifications.

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
ct= ColumnTransformer([("Town", OneHotEncoder(), [0])], remainder = 'passthrough')
```

```python
X
```

```python

X=ct.fit_transform(X)
```

```python

X
```

```python
X = X[:,1:]
X
```

```python
model1.fit(X,y)
```

```python
model1.predict([[1,0,2800]])
```

```python
model1.predict([[0,1,3400]])
```
